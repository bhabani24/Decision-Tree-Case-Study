```python
# DECISION TREE

# PLOTTING DECISION TREE

import os
import pandas as pd
import matplotlib.pyplot as plt


from sklearn import tree
from sklearn.tree import DecisionTreeClassifier

```

```python
os.chdir('D:/MLP_Session_26_JULY/DATASET')

iris = pd.read_csv('iris1.csv')
```

```python
iris
```

```python
# Creating different sets for Independent and Dependent variables

X = iris.iloc[:,:-1]
Y = iris.iloc[:,-1]
```

```python
iris.species.value_counts()
```

```python
cl=tree.DecisionTreeClassifier(random_state=123)
```

```python
# Fitting the model

cl=cl.fit(X,Y)
```

```python
plt.figure(figsize=(15,10))
tree.plot_tree(cl,filled=True)
```

```python
print(tree.export_text(cl))
```

```python
# PLOTTING DECISION TREE IN RANDOM FOREST

import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn import tree

from sklearn.ensemble import RandomForestClassifier
```

```python
cl=RandomForestClassifier(n_estimators=15, random_state=123)
```

```python
cl=cl.fit(X,Y)
```

```python
cl.estimators_
```

```python
len(cl.estimators_)
```

```python
plt.figure(figsize=(15,10))
tree.plot_tree(cl.estimators_[2],filled=True)
```

```python
plt.figure(figsize=(15,10))

for i in range(len(cl.estimators_)):
    tree.plot_tree(cl.estimators_[i],filled=True)

for i in range(len(cl.estimators_)):
    print(tree.export_text(cl.estimators_[i]))
```
